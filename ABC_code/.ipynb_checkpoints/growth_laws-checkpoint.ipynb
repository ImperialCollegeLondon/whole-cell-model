### Monods growth law
As per Weisse paper, Monod's law holds true:

if the steady-state synthesis time is approximately constant across different nutrient conditions and so the translational elongation rate, gamma, is approximately constant

* measure max and min gamma from a log range of external nutrient conditions

or

if enzyme activity is growth limiting.
